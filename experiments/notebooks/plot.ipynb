# Byol experiments

```python
path = r'C:\Users\tobih\OneDrive\Desktop\EigenspaceFiles\BYOLAffineFalseNoReg'
with open(os.path.join(path, 'F_eigenval.pkl'), 'rb') as f:
    F_eigenval = pickle.load(f)
with open(os.path.join(path, 'wp_eigenval.pkl'), 'rb') as f:
    wp_eigenval = pickle.load(f)
with open(os.path.join(path, 'allignment.pkl'), 'rb') as f:
    allignment = pickle.load(f)
with open(os.path.join(path, 'symmetry.pkl'), 'rb') as f:
    symmetry = pickle.load(f)

path = r'C:\Users\tobih\OneDrive\Desktop\EigenspaceFiles\BYOLAffineFalseReg'
with open(os.path.join(path, 'F_eigenval.pkl'), 'rb') as f:
    F_eigenval_reg = pickle.load(f)
with open(os.path.join(path, 'wp_eigenval.pkl'), 'rb') as f:
    wp_eigenval_reg = pickle.load(f)
with open(os.path.join(path, 'allignment.pkl'), 'rb') as f:
    allignment_reg = pickle.load(f)
with open(os.path.join(path, 'symmetry.pkl'), 'rb') as f:
    symmetry_reg = pickle.load(f)
```

```python
SMALL_SIZE = 10
MEDIUM_SIZE = 14
BIGGER_SIZE = 18

plt.rc('font', size=MEDIUM_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
fig, axs = plt.subplots(2, 4)
fig.set_size_inches(25, 10)
vis.eigenvalues_F(F_eigenval, axs[0,0])
vis.eigenvalues_Wp(wp_eigenval, axs[0,1])
vis.eigenspace_allignment(allignment, F_eigenval, axs[0,2])
vis.symmetry(symmetry, axs[0,3])
vis.eigenvalues_F(F_eigenval_reg, axs[1,0], title=False)
vis.eigenvalues_Wp(wp_eigenval_reg, axs[1,1], title=False)
vis.eigenspace_allignment(allignment_reg, F_eigenval, axs[1,2], title=False)
vis.symmetry(symmetry_reg, axs[1,3], title=False)
```
